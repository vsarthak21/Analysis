```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import mutual_info_score, roc_auc_score
from xgboost.sklearn import XGBRegressor
import scipy.stats as st
from sklearn.model_selection import GridSearchCV
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import mutual_info_score, roc_auc_score

```

```python
df = pd.read_excel("2010 Federal STEM Education Inventory Data Set.xls",skiprows=1)
df.head(2)
```

```python
def classifier(X_train, X_test, y_train, y_test):
    clf_dt = tree.DecisionTreeClassifier()
    clf_dt.fit(X_train, y_train)
  
    predicted_y = clf_dt.predict(X_test)
    clf_dt.score(X_test, y_test)

    mutual_score = mutual_info_score(y_test, predicted_y)
    roc_score = roc_auc_score(y_test, predicted_y)

    print('mutual score = ',mutual_score)
    print('roc_auc_score = ', roc_score)
    return mutual_score
```

```python
def xg( X_train, X_test, y_train, y_test):
    c = {  
      'max_depth': [5,10,15],
      'n_estimators': [50,100,150]
    }

    reg = XGBRegressor() 

    gs = GridSearchCV(reg, c, n_jobs=2)  
    gs.fit(X_train, y_train)  
    gs.best_params_
  
    xgb_predict = gs.predict(X_test)
    xg_pre = []
    for i in range(len(xgb_predict)):
        if xgb_predict[i]>0.49:
            xg_pre.append(1)
        else:
            xg_pre.append(0)
      
    xg_pre = np.array(xg_pre)
    score = gs.score(X_test, y_test)
    xg_roc = roc_auc_score(y_test, xg_pre)
    print('xgboost : ', score)
    print('xgboost roc :', xg_roc)
    return xg_roc
```

```python
df["C1) Funding FY2008"].fillna(0.0,inplace = True)
df["C2) Funding FY2009"].fillna(0.0,inplace = True)
```

```python
df = df[df["C1) Funding FY2008"] != 0.0]
df = df[df["C2) Funding FY2009"] != 0.0]
```

```python
df["%_growth"]=((df["C2) Funding FY2009"] - df["C1) Funding FY2008"]) / df["C1) Funding FY2008"])*100
```

```python
def get_labels(n):
    if n >= 0:
        return 1
    else:
        return 0
df["labels"] = df["%_growth"].apply(get_labels)
```

```python
f_cols = ['C1) Funding FY2008', 'C2) Funding FY2009', 'C3) Funding FY2010']
nf_cols = df0.drop(f_cols, axis = 1)

cols = nf_cols.columns[nf_cols.isnull().sum() < 60]
df_nf = nf_cols[cols]

df_nf
```

```python
funds = df.filter(['C1) Funding FY2008','C2) Funding FY2009'],axis = 1)
growth = df.filter(['%_growth','labels'],axis = 1)
```

```python
df_ = df.drop(df.columns[:9], axis=1)
df_ = df_.drop(['%_growth','labels'], axis=1)
df_cat = pd.DataFrame({col: df_[col].astype('category').cat.codes for col in df_}, index=df_.index)
df_cat.head(3)
```

```python
final_data = pd.concat([df_cat,funds,growth],axis=1)
final_data.head(3)
```

```python
final_data = final_data.dropna(how = 'all')
```

```python
data_points = final_data.values[:,:250]
labels = final_data.values[:,-1]
```

```python
Xtrain,Ytrain,Xtest,Ytest = train_test_split(data_points,labels,test_size=0.30,random_state = 1)
```

```python
classifier(Xtrain,Ytrain,Xtest,Ytest)
xg(Xtrain,Ytrain,Xtest,Ytest)
```
